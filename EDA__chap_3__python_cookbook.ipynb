## **Create new column Sepal Area and Petal Area using available variables**

---



```python
#Multiplying three different columns in order to check wether i can perform operations on more than two columns
df['sepal_Area']=df['sepal_length']*df['sepal_width']/df['petal_length']
df['petal_Area']=df['petal_length']*df['petal_width']
df.head()
```

```python
#It returns only the rows where sepal length is greater than sepal witdth
df[df['sepal_length'] > df['sepal_width']]
```

# Sorted data on basis of Sepal Area

```python
#Data set showing sorted data with largest sepal area to smallest and vice versa
df=df.sort_values('sepal_Area',ascending=False)
df.head()
```

```python
df=df.sort_values('sepal_Area',ascending=True)
df.head()
```

## Sorted data set on basis of both sepal length and peta length
This command means:

First, sort by sepal_length in descending order
Then, within each group of identical sepal_length values, sort by petal_length in ascending order.

```python
df.sort_values(['sepal_length','petal_length'],ascending=False)
```

I organized the flower data by two things:

 First, I looked at the length of the sepal (the outer part of the flower) — I put the longest ones at the top.

Then, if two flowers had the same sepal length, I compared their petal length — and listed the one with the shorter petal first.

So overall, I grouped flowers by how long their sepals are, and for flowers with the same sepal size, I sorted them further by petal size.

```python
df.sort_values(['sepal_length','petal_length'],ascending=[False,True])
```

# Line Plot without Species (Simple Line Plot)
This line plot shows the relationship between the length of the sepal and the length of the petal in iris flowers. As the sepal length increases, the petal length generally increases too — the line trends upward, which means there’s a positive correlation between these two features. The graph helps us understand that larger sepals are usually associated with larger petals, regardless of the flower species.

```python
sns.lineplot(data=df, x='sepal_length', y='petal_length')

plt.title('Petal Length vs Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

This colored line plot helps compare trends across species visually.

```python
sns.lineplot(data=df, x='sepal_length', y='petal_length', hue='species')

plt.title('Petal Length vs Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

# Plot a bar chart using GGPLOT
The bar chart displays the average petal width for each species of the Iris flower: Setosa, Versicolor, and Virginica. From the chart, we can observe that:

Setosa has the smallest average petal width, around 0.25 cm.

Versicolor has a moderate average of about 1.33 cm.

Virginica has the largest average petal width, approximately 2.03 cm.

This visualization clearly highlights how petal width increases from Setosa to Virginica, indicating that petal width is a good feature to help differentiate species in the Iris dataset.

```python
grouped = df.groupby('species')['petal_width'].mean().reset_index()

plot = (
    ggplot(grouped, aes(x='species', y='petal_width')) +
    geom_bar(stat='identity', fill='skyblue', color='black') +
    labs(
        title='Average Petal Width by Iris Species',
        x='Species',
        y='Petal Width (cm)'
    ) +
    theme(figure_size=(12, 6))
)

display(plot)
```

# **Bokeh(BarChart)**
"I used Bokeh to create a bar chart that shows how many samples belong to each Iris species in the dataset.

The three bars represent Setosa, Versicolor, and Virginica.

The height of each bar shows how many (flowers) are present for that species.

This helps us quickly see if the dataset is balanced or if one species has more/less data.

I used Bokeh because the chart is clean and interactive (zoom, pan, etc.), and gives a good first visual summary of the dataset."



```python

df['species'] = df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

species_counts = df['species'].value_counts().sort_index()
species = list(species_counts.index)
counts = list(species_counts.values)


fig = figure(x_range=species, title="Iris Species Count", width=800, height=400)


fig.vbar(x=species, top=counts, width=0.5)


show(fig)
```
